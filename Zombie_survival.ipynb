```python
import time
import random

def start_adventure():
    print("🧟 VIRTIXEL: ZOMBIE SURVIVAL 🧟")
    time.sleep(1)
    print("\nYou are standing on a dark highway. You hear groaning nearby.")

    supplies = ["FASHLIGHT", "BANDAGE"]
    print(f"Your starting supplies: {supplies}")

    choice1 = input("\nDo you want to hide in the GAS STATION or run to the WOODS? ").upper()

    if "GAS" in choice1 or "STATION" in choice1:
        print("\nYou sneak into the gas station. It's quiet...")
        time.sleep(1.5)
        # Random chance element
        if random.choice([True, False]):
            print("🎁 Jackpot! You found a Shotgun and escaped the horde! YOU WIN!")
        else:
            print("☠️ Oh no! A zombie was hiding behind the counter. Game Over!")

    elif "WOODS" in choice1 or "RUN" in choice1:
        print("\nYou run into the dark woods and trip over a log.")
        time.sleep(1.5)
        print("🌲 The groaning fades away. You are safe for now... YOU SURVIVED!")

    else:
        print("\nYou froze in fear! The zombies caught you. Game Over!")

# Run the project
start_adventure()
```
