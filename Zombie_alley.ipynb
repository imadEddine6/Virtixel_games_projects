```python
import random
import time

def start_game():
    print("🧟 Welcome to ZOMBIE ALLEY! 🧟")
    print("---------------------------------")
    time.sleep(1)
    print("You are standing in a dark alleyway. You hear growling nearby.")
    print("You have two choices:")

    # Game Loop
    game_running = True
    while game_running:
        choice = input("\nDo you want to (1) Hide behind a trash can or (2) Run to the main street? Enter 1 or 2: ")

        if choice == "1":
            print("\nYou hide quietly. A zombie walks right past you...")
            time.sleep(1.5)
            print("Lucky! The zombie didn't see you. You found a safe house and escaped!")
            print("🎉 YOU WIN! 🎉")
            game_running = False
        elif choice == "2":
            print("\nYou sprint toward the lights of the main street!")
            time.sleep(1.5)
            # 50/50 chance of survival
            if random.choice([True, False]):
                print("You outran them! You made it to the military safe zone.")
                print("🎉 YOU WIN! 🎉")
            else:
                print("Oh no! A pack of zombies cut you off. You became dinner.")
                print("💀 GAME OVER! 💀")
            game_running = False
        else:
            print("Invalid choice! Fear makes you freeze up. Try again quickly!")

# Launch the game
if __name__ == "__main__":
    start_game()
```
